# ANALYSIS OF EUROPEAN-STYLE CALL AND PUT OPTIONS

In this notebook I show how you can analyse Vanilla European Call and Put Equity options

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from financepy.finutils import *
from financepy.products.equity import *
from financepy.market.curves import *
```

## Creating the Option

We need to set up the option details - what you would find in the Term Sheet.

```python
expiryDate = TuringDate(1, 7, 2015)
strikePrice = 100.0
```

We now create the option object

```python
callOption = TuringEquityVanillaOption(expiryDate, strikePrice, TuringOptionTypes.EUROPEAN_CALL)
```

```python
print(callOption)
```

```python
putOption = TuringEquityVanillaOption(expiryDate, strikePrice, TuringOptionTypes.EUROPEAN_PUT)
```

## Valuation

To do a valuation we need to specify the date and set the market inputs

```python
valueDate = TuringDate(1, 1, 2015)
stockPrice = 90.0
dividendYield = 0.0
```

```python
dividendCurve = TuringDiscountCurveFlat(valueDate, dividendYield)
```

For simplicity the discount curve is assumed to be flat here (we can pass in any discount curve) and is created as follows.

```python
interestRate = 0.02
discountCurve = TuringDiscountCurveFlat(valueDate, interestRate, TuringFrequencyTypes.ANNUAL)
```

The valuation model we use is Black-Scholes which is passed in as a Black-Scholes model object. It is created as follows.

```python
volatility = 0.20
model = TuringModelBlackScholes(volatility)
```

The valuation is as follows and the output is in a dictionary

```python
callValue = callOption.value(valueDate, stockPrice, discountCurve, dividendCurve, model)
print("Call Option Value: %12.5f" %callValue)
```

```python
putValue = putOption.value(valueDate, stockPrice, discountCurve, dividendCurve, model)
print("Put Option Value: %12.5f" %putValue)
```

### Spot Price Dependency

We can examine how the option value depends on the spot price

```python
stockPrices = np.linspace(60,140,100)
values = callOption.value(valueDate, stockPrices, discountCurve, dividendCurve, model)
```

```python
plt.figure(figsize=(10,6))
plt.plot(stockPrices,values, label="Call Option Prices")
plt.xlabel("Stock Prices")
plt.ylabel("Value")
plt.legend()
plt.grid()
```

### Volatility Dependency

We can see how the value depends on volatility

```python
volatilities = np.linspace(0.01,10,100)
models = TuringModelBlackScholes(volatilities)
values = callOption.value(valueDate, stockPrice, discountCurve, dividendCurve, models)
```

```python
plt.figure(figsize=(10,6))
plt.plot(volatilities*100,values, label="Option Prices")
plt.xlabel("Volatility (%)")
plt.ylabel("Value")
plt.legend()
plt.grid()
```

### Strike Price Dependency

Also strikes. 

```python
strikePrices = np.linspace(70,130,50)
```

But as these are part of the option description - so we need to first vectorise those.

```python
callOptions = TuringEquityVanillaOption(expiryDate, strikePrices, TuringOptionTypes.EUROPEAN_CALL)
```

Look, we have an array of callOptions with different strikes

```python
callOptions._strikePrice
```

Then we can get back a vector of values

```python
callValues = callOptions.value(valueDate, stockPrice, discountCurve, dividendCurve, model)
```

```python
plt.figure(figsize=(10,6))
plt.plot(strikePrices, callValues, label="CALL Option Prices")
plt.xlabel("Strike Price")
plt.ylabel("Value")
plt.legend()
plt.grid()
```

Let's do put options too 

```python
putOptions = TuringEquityVanillaOption(expiryDate, strikePrices, TuringOptionTypes.EUROPEAN_PUT)
```

```python
putValues = putOptions.value(valueDate, stockPrice, discountCurve, dividendCurve, model)
```

```python
plt.figure(figsize=(10,6))
plt.plot(strikePrices, putValues, label="PUT Option Prices")
plt.xlabel("Strike Price")
plt.ylabel("Value")
plt.legend()
plt.grid()
```

## Risk Sensitivities

```python
callDeltas = callOption.delta(valueDate, stockPrices, discountCurve, dividendCurve, model)
```

```python
plt.figure(figsize=(10,6))
plt.plot(stockPrices, callDeltas, label="CALL Delta")
plt.xlabel("Stock Price")
plt.ylabel("Value")
plt.legend()
plt.grid()
```

```python
callGammas = callOption.gamma(valueDate, stockPrices, discountCurve, dividendCurve, model)
```

```python
plt.figure(figsize=(10,6))
plt.plot(stockPrices, callGammas, label="CALL Gamma")
plt.xlabel("Stock Price")
plt.ylabel("Value")
plt.legend()
plt.grid()
```

```python
callVegas = callOption.vega(valueDate, stockPrices, discountCurve, dividendCurve, model)
```

```python
plt.figure(figsize=(10,6))
plt.plot(stockPrices, callVegas, label="CALL Vega")
plt.xlabel("Stock Price")
plt.ylabel("Value")
plt.legend()
plt.grid()
```

```python
callThetas = callOption.theta(valueDate, stockPrices, discountCurve, dividendCurve, model)
```

```python
plt.figure(figsize=(10,6))
plt.plot(stockPrices, callThetas, label="CALL Theta")
plt.xlabel("Stock Price")
plt.ylabel("Value")
plt.legend()
plt.grid()
```
